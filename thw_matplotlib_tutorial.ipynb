# ``pyplot`` vs Object-oriented 

There are two basic methods for using the matplotlib library

### The ``pyplot`` api
- High-level function calls that automatically handle most details
- Similar to the Matlab plotting
- Good for simple, quick plots, and those requiring only slight modifications to default settings.

[pyplot tutorial](http://matplotlib.org/users/pyplot_tutorial.html)

### Object-oriented structure
- access to control of all underlying structure of the figure object
- usually more verbose than ``pyplot``

The two methods can often be used interchangeably on in the same code

## A brief aside on "backends"
Matplotlib has multiple backends, the program that handle the graphical interface. These allow matplotlib to interface with a variety of GUI toolkits (GTK, Qt, Wx, etc.), and therefor to behave the same (most of the time) on different platforms. 

In most cases, it should make no difference which of backend matplotlib uses. However, bugs do occur, and so it is often useful to know which version of matplotlib, and which backend you are using.

```python
import matplotlib as mpl

print(mpl.__version__)
print(mpl.get_backend())
```

To start with we enable ``pylab`` support in the ipython notebook. Omit this if you are using a .py script.

In python you import different modules in order to gain access to functionality beyond the core python langauge.

The following imports ``matplotlib``, ``pyplot``, as well as ``numpy`` *the* scientific computing package in python, which is based on array objects.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

First create some arrays of data using some basic ``numpy`` functions.

```python
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)
z = np.cos(x)
```

Using pyplot it is only a few lines create plots.

```python
# Generate a plot of the arrays
plt.plot(x,y, label='sin(x)')

# Add a second line to that plot
plt.plot(x, z, label='cos(x)')

# label the plot and add a legend
plt.title('Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Note: if your script is not displaying the plot you need to call ``plt.show()`` at the end of the script. This is being done automatically by ipython.

Here is the same example, written to explicitly show the object-orientated structure of the figure

```python
f, ax0 = plt.subplots()                  # we manually make a figure and axis

ax0.plot(x, y, label='sin(x)')           # it's the axis who plots
ax0.plot(x, z, label='cos(x)')  

ax0.legend()
ax0.set_title('Trigonometric Functions') # we set the title on the axis
ax0.set_xlabel('x')             
ax0.set_ylabel('y')
```

We now have explicit references to the figure object ``f`` and the axis object ``ax0`` that it contains.

Whereas ``pyplot`` typically adds new elements to the last figure and axis used, the object-oriented notation specifies exactly where you are putting the element. This can help avoid confusion when working on multiple plots, or one plot with multiple subplots.

Note that ``pyplot`` is using the same objects in the background, and you can acces the current axis and figure using.

    plt.gcf()  # get current figure
    plt.gca()  # get current axis

#Making subplots

Subplots can be made using the command

    f, (ax1,ax2,...) = plt.subplots(n,m)

which is a shorter version of the the more explicit:

    f = plt.figure()
    ax1 = f.add_subplot(n, m, 1)
    ax2 = f.add_subplot(n, m, 2)
    ...

With the arguments ``(n,m,s)`` we are laying out an ``n`` row by ``m`` column grid of subplots, and specifying the ``s``th subplot.

```python
#%matplotlib 

x = np.linspace(0.01, 8 * np.pi, 400)
y = np.sin(x)*x
z = np.cos(x)/(x)*30.

# Two subplots, unpack the output array immediately
#f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
f, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(x, y)
ax2.plot(x, z)

# Put a figure-level title
f.suptitle('Subplots')
# put a title for a subplot
ax2.set_title('Second Subplot')
```

Note that, by default, matplotlib is choosing the scale on the two subplots independently. 

We can instead choose our own scale, and have them share the same y-axis.

```python
# Two subplots, unpack the output array immediately
f = plt.figure()
ax1 = f.add_subplot(1, 2, 1)
ax1.scatter(x, y)

ax2 = f.add_subplot(1,2,2,sharex=ax1,sharey=ax1)    # Plot second function with the same y-scale
plt.setp( ax2.get_yticklabels(), visible=False) # Make tick labels appear only for the 1st subplot
ax2.plot(x, z)

ax1.set_ylim(-30,30)   # set the desired upper and lower limit to be shown on the y-axis

# Adjust the space between subplots
f.subplots_adjust(wspace=0.05)

# Put a figure-level title
f.suptitle('Shared Y-axis')
```

Another function, subplot2grid, can be used to generate arbitrary arrangements of subplots

```python
f = plt.figure()

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))

# Turn off tick labels
for ax in f.axes:
    ax.set_xticks([])
    ax.set_yticks([])
```

# Lines, collections, patches etc...

The next objects in the matplotlib hierachy, are representations of the elements plotted on the axis. These can all be accessed to modify their properties in an existing plot.

Updating the plot is done through the figure.canvas object.

For example we can access elements in our previous two plots. In the line plot, the objects of interest are lines.

```python
print ax_lines.lines

aline = ax_lines.lines[0]
aline.set_linewidth(5)
f_lines.canvas.draw()
```

```python
col = ax_scatter.collections[0]
col.set_color('r')
f_scatter.canvas.draw()
```

```python
col.set_visible(False)
f_scatter.canvas.draw()
```

These objects also keep track of their positions on the plot, which can be modified.

```python
print aline.get_xdata()
print aline.get_ydata()
```

```python
x = np.linspace(0.,10.,10)
y = x**2
aline.set_data(x,y)
aline.get_ydata()
```

# Common plot types

Here are some examples of common plot types. These examples are taken directly from Joshua Blooms course notes.

Error plots
-----------

First a very simple error plot

```python
%matplotlib inline
# example data
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

# example variable error bar values
yerr = 0.1 + 0.2 * np.sqrt(x)
xerr = 0.1 + x / 15

# First illustrate basic pyplot interface, using defaults where possible.
plt.figure()
plt.errorbar(x, y, xerr=xerr, yerr=yerr, fmt='o', ecolor='g')
```

Logarithmic plots
-----------------

```python
x = np.linspace(-5, 5)
y = 2 ** x

f, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(x, y)
ax2.semilogy(x, y)
```

Bar plots
---------

```python
N = 5
men_means = (20, 35, 30, 31, 27)
men_std = (2, 3, 4, 1, 2)
women_means = (25, 32, 34, 21, 29)
women_std = (3, 5, 2, 3, 3)

ind = np.arange(N)  # the x locations for the groups
width = 0.35        # the width of the bars

fig, ax = plt.subplots()

rects1 = ax.bar(ind, men_means, width, color='r', yerr=men_std)
rects2 = ax.bar(ind + width, women_means, width, color='y', yerr=women_std)

# add some
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((rects1, rects2), ('Men', 'Women'))
```

Scatter plots
-------------

The ``scatter`` command produces scatter plots with arbitrary markers.

```python
from matplotlib import cm    # cm == colormap

t = np.linspace(0.0, 6*np.pi, 100)
y = np.exp(-0.1 * t) * np.cos(t)
phase = t % (2 * np.pi)

f, ax = plt.subplots()
ax.scatter(t, y, s=100 * abs(y), c=phase, cmap=cm.jet)
ax.set_ylim(-1, 1)

ax.grid()
ax.axhline(color='black')
```

Note that scatter plots allow you to make marker size and color a function of the data itself.

Histograms
----------

Matplotlib has a built-in command for histograms.

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.25)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

# Customizing matplotlib's default (RC) paramers

Matplotlib has an extensive set of default parameters for almost every property of a plot.

These are called the rc settings or rc parameters and the defaults are stored in a configuration file:

```python
# Print the location of the rc parameters
mpl.matplotlib_fname()
```

Changing the parameters in this file will change the default properties of all matplotlib plots.

This is useful when generating a number of different plots, with the same settings. For instance, making figures for a publication with specific formatting requirements. Setting the rc parameters avoids the need to repeatedly reset the same attributes for each plot.

## Dynamic modification of rc parameters

In addition to modifying the the matplotlibrc file, these parameters can also be modified on a script- (or notebook-) wide scope.

They are stored locally in a `dict` object...

```python
mpl.rcParams
```

```python
isinstance(mpl.rcParams,dict)
```

... in which you can update particular properties directly.

```python
mpl.rcParams['axes.edgecolor'] = 'r'
mpl.rcParams['axes.facecolor'] = 'y'

plt.plot([1,2,3])
```

Your can also set multiple properties at once using the `rc()` function.

```python
mpl.rc('lines', linewidth=2, color='r')
```

It also handy to be able to reset these back to matplotlib's internal defaults, or to the configuration file defaults.

```python
# reset rcparams to internal defaults
mpl.rcdefaults()

# reset rcparams to configuration file
mpl.rc_file('matplotlibrc')

plt.plot([1,2,3])
```

Note also that the ipython "magicks" update these parameters.

```python
%matplotlib inline

plt.plot([1,2,3])
```

## Example: "Dark theme" presentation

Suppose you wanted to give a mpl based presentation in an ipython notebook, but the default theme just isn't snazzy enough for you.

We can instead render it with a different theme, i.e.

    ipython notebook --profile=default-dark

Now you will notice that the default matplotlib settings are meant for a light background.

```python
# rcParams for a dark background
mpl.rc('lines',color='w') 
#mpl.rc('patch',edgecolor='w')
mpl.rc('text',color='w')
mpl.rc('axes',facecolor='k',edgecolor='w',labelcolor='w')
mpl.rc('xtick',color='w')
mpl.rc('ytick',color='w')
#mpl.rc('figure',facecolor='k',edgecolor='k')
#mpl.rc('savefig',facecolor='k',edgecolor='k')
```

Now lets try generating the first plot in this presentation with the new defaults.

```python
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)
z = np.cos(x)

f2, ax0 = plt.subplots()                  # we manually make a figure and axis

ax0.plot(x, y, label='sin(x)')           # it's the axis who plots
ax0.plot(x, z, label='cos(x)')  

ax0.legend()
ax0.set_title('Trigonometric Functions') # we set the title on the axis
ax0.set_xlabel('x')             
ax0.set_ylabel('y')
plt.grid(color='w')
```

Now the primary colors on a dark background looks a little strong. These colors are stored in a list called the color cycle.

```python
mpl.rcParams['axes.color_cycle']
```

We can change these to a default parameter that looks nicer with the dark background

```python
# Muted color cycle, colors listed in hex
hex_cols = [u'#4878cf', u'#6acc65', u'#d65f5f', u'#b47cc7', u'#c4ad66', u'#77bedb']
mpl.rc('axes',color_cycle=hex_cols)
```

I cheated a little bit (I cant read colors in hex either). I generate these using seaborn.
    
    import seaborn as sns
    cols = sns.color_palette('muted')
    hex_cols = [mpl.colors.rgb2hex(col) for col in cols]
    mpl.rcParams['axes.color_cycle'] = hex_cols

Now remaking the same plot.

```python
f2, ax0 = plt.subplots()                  # we manually make a figure and axis

ax0.plot(x, y, label='sin(x)')           # it's the axis who plots
ax0.plot(x, z, label='cos(x)')  

ax0.legend()
ax0.set_title('Trigonometric Functions') # we set the title on the axis
ax0.set_xlabel('x')             
ax0.set_ylabel('y')
plt.grid(color='w')
```

Note that some plot types (such as scatter) dont use the color cycle variable.
