# ``pyplot`` vs Object-oriented 

There are two basic methods for using the matplotlib package.

### The ``pyplot`` api
- High-level function calls that automatically handle most details
- Similar to the Matlab plotting
- Good for simple, quick plots, and those requiring only slight modifications to default settings.

[pyplot tutorial](http://matplotlib.org/users/pyplot_tutorial.html)

### Object-oriented structure
- access to control of all underlying structure of the figure object
- usually more verbose than ``pyplot``
- can build more complex functionality into your figures

The two methods can usually be used interchangeably in the same code, since they are two different ways of working with the same python objects.

## A brief aside on "backends"
Matplotlib supports multiple backends, the programs that handle the graphical representation of the figure. These allow matplotlib to interface with a variety of GUI toolkits (GTK, Qt, Wx, etc.), and therefor to behave the same (most of the time) on different platforms, since any given computer will use different toolkit from another. 

In most cases, it should make no difference which of backend matplotlib uses. However, bugs do occur, and so it is often useful to know which version of matplotlib, and which backend you are using.

```python
import matplotlib as mpl

print(mpl.__version__)
print(mpl.get_backend())
```

#Making subplots

Subplots can be made using the command

    f, (ax1,ax2,...) = plt.subplots(n,m)

which is a shorter version of the the more explicit:

    f = plt.figure()
    ax1 = f.add_subplot(n, m, 1)
    ax2 = f.add_subplot(n, m, 2)
    ...

With the arguments ``(n,m,s)`` we are laying out an ``n`` row by ``m`` column grid of subplots, and specifying the ``s``th subplot.

```python
x = np.linspace(0.01, 8 * np.pi, 400)
y = np.sin(x)*x
z = np.cos(x)/(x)*30.

# Two subplots
f, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(x, y)
ax2.plot(x, z)

# Put a figure-level title
f.suptitle('Subplots')
# put a title for a subplot
ax2.set_title('Second Subplot')
```

Note that, by default, matplotlib is choosing the scale on the two subplots independently. And this might not be the most useful for comparing the two functions.

We can instead choose our own scale, and have them share the same y-axis.

```python
f = plt.figure()
ax1 = f.add_subplot(1, 2, 1)
ax1.scatter(x, y)

ax2 = f.add_subplot(1,2,2,sharex=ax1,sharey=ax1)    # Plot second function with the same y-scale
plt.setp( ax2.get_yticklabels(), visible=False) # Make tick labels appear only for the 1st subplot
ax2.plot(x, z)

ax1.set_ylim(-30,30)   # set the desired upper and lower limit to be shown on the y-axis

# Adjust the space between subplots
f.subplots_adjust(wspace=0.05)

# Put a figure-level title
f.suptitle('Shared Y-axis')
```

#Decorations

Here is our same plot of the sinusoidal functions displaying a few additional modifications that can be made to improve the look of a plot in just a few lines.

```python
f, ax = plt.subplots()

# Three simple polyniomials
x = np.linspace(-2*np.pi,2*np.pi)
y = np.sin(x)
z = np.cos(x)

# Plot each with a TeX labels (for a legend)
ax.plot(x/np.pi, y, label=r'$\sin(x)$') 
ax.plot(x/np.pi, z, label=r'$\cos(x)$')

# Make all lines drawn so far thicker
plt.setp(ax.lines, linewidth=2)

# set limits
ax.set_xlim(-2,2.01)
ax.set_ylim(-1.01,1)

# fill space between lines
ax.fill_between(x/np.pi, y, z,alpha=0.3) # alpha controlls transparency

# manually set tick mark spacing and labels
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

majorLocator   = MultipleLocator(1.)
majorFormatter = FormatStrFormatter(r'%d $\pi$')
minorLocator   = MultipleLocator(.25)

majorLocator   = MultipleLocator(0.5)

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_major_formatter(majorFormatter)
ax.xaxis.set_minor_locator(minorLocator)

# Add a grid and a legend that doesn't overlap the lines
ax.grid()
ax.legend(loc='lower right')

# Add black horizontal and vertical lines through the origin
ax.axhline(0, color='black')
ax.axvline(0, color='black')

# No frame around the axes
ax.set_frame_on(False)

# Set main text elements of the plot with TeX symbols
ax.set_title('Sinusoids')
ax.set_xlabel(r'$x$ (radians)')
ax.set_ylabel(r'$p(x)$')
```

#Text and annotations

You can add text at arbitrary locations on your figures. Here are some examples for how this text can be formated.

Note that the location of the text uses the same coordinate system as the the plotted data.

# Common plot types

Error plots
-----------

First a simple error plot.

```python
%matplotlib inline
# example data
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

# example variable error bar values
yerr = 0.1 + 0.2 * np.sqrt(x)
xerr = 0.1 + x / 15

# First illustrate basic pyplot interface, using defaults where possible.
plt.figure()
plt.errorbar(x, y, xerr=xerr, yerr=yerr, fmt='o', ecolor='g')
```

Logarithmic plots
-----------------

``semilogx``, ``semilog`` and ``loglog`` functions.

```python
x = np.linspace(-5, 5)
y = 2 ** x

f, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(x, y)
ax2.semilogy(x, y)
```

Bar plots
---------

Bar plots with multiple data sources and error bars.

```python
N = 5
men_means = (20, 35, 30, 31, 27)
men_std = (2, 3, 4, 1, 2)
women_means = (25, 32, 34, 21, 29)
women_std = (3, 5, 2, 3, 3)

ind = np.arange(N)  # the x locations for the groups
width = 0.35        # the width of the bars

fig, ax = plt.subplots()

rects1 = ax.bar(ind, men_means, width, color='r', yerr=men_std)
rects2 = ax.bar(ind + width, women_means, width, color='y', yerr=women_std)

# add some
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((rects1, rects2), ('Men', 'Women'))
```

Scatter plots
-------------

The ``scatter`` command produces scatter plots with arbitrary markers. Note here that size and color are set by the data values. Values are mapped onto color using a colormap.

```python
from matplotlib import cm    # cm == colormap

t = np.linspace(0.0, 6*np.pi, 100)
y = np.exp(-0.1 * t) * np.cos(t)
phase = t % (2 * np.pi)

f, ax = plt.subplots()
ax.scatter(t, y, s=100 * abs(y), c=phase, cmap=cm.jet)
ax.set_ylim(-1, 1)

ax.grid()
ax.axhline(color='black')
```

Note that scatter plots allow you to make marker size and color a function of the data itself.

# Customizing matplotlib's default (RC) paramers

Matplotlib has an extensive set of default parameters for almost every property of a plot.

These are called the rc settings or rc parameters and the defaults are stored in a configuration file (just like you would for vim or bash):

```python
# Print the location of the rc parameters
%matplotlib inline
mpl.matplotlib_fname()
```

I have example matplotlibrc file in the same directory called ``matplotlibrc_example``. Try sourcing the matplotlibrc locally.

    mv matplotlibrc_example matplotlibrc

Changing the parameters in this file will change the default properties of all matplotlib plots.

This is useful when generating a number of different plots, with the same settings. For instance, making figures for a publication with specific formatting requirements. Setting the rc parameters avoids the need to repeatedly reset the same attributes for each plot.

## Dynamic modification of rc parameters

In addition to modifying the the matplotlibrc file, these parameters can also be modified on a script- (or notebook-) wide scope.

They are stored locally in a `dict` object...

```python
mpl.rcParams
```

```python
isinstance(mpl.rcParams,dict)
```

... in which you can update particular properties directly.

```python
mpl.rcParams['axes.edgecolor'] = 'r'
mpl.rcParams['axes.facecolor'] = 'y'

plt.plot([1,2,3])
```

Your can also set multiple properties at once using the `rc()` function, which uses a keyword to determine what objects you are modifying the properties.

```python
mpl.rc('lines', linewidth=4, linestyle='--')
```

It also handy to be able to reset these back to matplotlib's internal defaults, or to the configuration file defaults.

```python
# reset rcparams to internal defaults
mpl.rcdefaults()

# reset rcparams to configuration file
#mpl.rc_file('matplotlibrc')

plt.plot([1,2,3])
```

#Toolkits

http://matplotlib.org/mpl_toolkits/

These are additional packages that provide more specific functionality that is not bundled with the main matplotlib package. (They must be downloaded separately).

These include packages for mapping, 3d plotting, interfacing with different data sources and various ways of modifying the look of plots. Seaborn is an example of the later.

```python
%matplotlib inline
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

# Draw the stations on a real map of the Earth.
# Find boundaries 
lon0 = -10.
lon1 = 30.
lat0 = 35.
lat1 = 70.

# Geographic grid to draw
parallels = np.linspace(lat0, lat1, 5)
meridians = np.linspace(lon0, lon1, 5)

# Resolution of the basemap to load ('f' is *very* expensive)
resolution = 'i' # intermediate resolution for map info

# Create the figure
f2, ax2 = plt.subplots(figsize=(10,6))

# Make the map object, where we draw geographic information
m = Basemap(lon0, lat0, lon1, lat1, resolution=resolution, ax=ax2)
m.drawcountries(color=(1,1,0))  # country boundaries yellow
m.drawrivers(color=(0,1,1))  # rivers in cyan
m.bluemarble()  # NASA bluemarble image
m.drawparallels(parallels, labels=[1,0,0,0], fmt='%.2f')
m.drawmeridians(meridians, labels=[0,0,0,1], fmt='%.2f')

# location of some cities 40.4000° N, 3.6833° W, 48.8567° N, 2.3508° E, 52.5167° N, 13.3833° E
names = ['Madrid','Paris','Berlin']
lats = np.array([40.4000,48.8567,52.5167])
longs = np.array([-3.6833,2.3508,13.3833])

# Now draw a scatter plot on the same map
s = m.scatter(longs,lats, c='r',s=50)

# Label the figure and every point by its station label
#ax2.set_title(title)
for lat,lon,name in zip(lats,longs,names):
    ax2.text( lon+0.5, lat+0.5, name, 
            weight='bold', color='yellow', zorder=10)
```

# Event Handling

Matplotlib provides an API for interacting with a figure via key presses and mouse movements. 

This is “GUI neutral” so you don’t have to repeat a lot of code across the different user interfaces. It's capabilities are limited, so for very complex or fast interactions you are better off choosing a specific Graphical User Interface (GUI) toolkit and using its specific event model. For many scientific uses, what matplotlib offers is more than sufficient.

Here is a very simple example in which matplotlib will record mouse clicks on an empty plot.

```python
#change some default parameters (this wont work inline.)
%matplotlib 
fig = plt.figure()
ax = fig.add_subplot(111)

def onclick(event):
    print 'button=%d, x=%d, y=%d, xdata=%f, ydata=%f'%(
        event.button, event.x, event.y, event.xdata, event.ydata)

cid = fig.canvas.mpl_connect('button_press_event', onclick)
```

The FigureCanvas method mpl_connect returns a connection id which is simply an integer. When you want to disconnect the callback, just call::
fig.canvas.mpl_disconnect(cid)
The most commonly used event types are KeyEvent and MouseEvent, both of which have the following attributes:
``x``: 
    x position - pixels from left of canvas

``y``: 
    y position - pixels from bottom of canvas

``inaxes``: 
    the ``matplotlib.axes.Axes`` instance if mouse is over axes

``xdata``: 
    x coord of mouse in data coords

``ydata``:
    y coord of mouse in data coords
In addition, MouseEvent have:
``button``
    button pressed None, 1, 2, 3, 'up', 'down' (up and down are used for
    scroll events)

``key``
    the key pressed: None, any character, 'shift', 'win', or 'control'

For a more complicated example I will grab one directly from the matplotlibs example website with the following code. You can also find this example by navigating to http://matplotlib.sourceforge.net/examples/event_handling/data_browser in a web browser.

This provides code for a point browser, which allows one to pick a data series to view based on its position on a plot of its average and standard deviation.
